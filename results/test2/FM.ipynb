```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
losses = np.genfromtxt('losses.csv', delimiter=',')
train_losses = losses[:, 0]
val_losses = losses[:, 1]

# Plot the training and validation loss
epochs = list(range(len(train_losses)))
plt.plot(epochs, train_losses, 'o-', label='Train loss', lw=2, color='black')
plt.plot(epochs, val_losses, 'o-', label='Val loss', lw=2, color='gray')
plt.xlabel('Epoch')
plt.ylabel('FM loss (MSE)')
plt.legend(loc = 'lower left', frameon=False)
plt.title(f'FM: Scale 16, {len(train_losses)} epochs, min val loss: {min(np.round(val_losses, 6))}')
plt.grid(True)
plt.xlim(-1, len(train_losses))
plt.yscale('log')
```
